```python
#Street Food Cart

menu = {"Egg Rice": 60,
        "Veg Rice": 50,
        "Gobi": 60,
        "Gobi Rice": 70,
        "Noodles": 60,
        "Gobi Noodles": 80,
        "Dry Gobi": 30,
        "Capsicum Gobi": 80}

cart = []
total = 0

print("-------------- MENU --------------")
for key, value in menu.items():
  print(f"{key:20} : ₹{value:.1f}")
print("------------------------------------------")

while True:
  food = input("Select an item (q to quit)")
  if food.lower() == "q":
    break
  elif menu.get(food) is not None:
    cart.append(food)

print("---------------YOUR ORDER-----------------")
for food in cart:
  total += menu.get(food)
  print(food, end=" ")

print()
print(f"Total is: ₹{total:.1f}")
```
