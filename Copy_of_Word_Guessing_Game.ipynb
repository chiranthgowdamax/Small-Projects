```python
secret_word = "Chad"
guess = ""
guessed_count = 0
trial = 3
out_of_guesses = False

while guess != secret_word and not(out_of_guesses):
  if guessed_count < trial:
   guess = input("Enter guess:")
   guessed_count += 1
  else:
    out_of_guesses = True

if out_of_guesses:
  print("Out of Guesses, You Lose!")
else:
  print("You win!")
```
