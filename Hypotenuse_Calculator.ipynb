```python
import math

print("--- HYPOTENUSE OF RIGHT TRIANGLE ---")
print(" ")
a = float(input("Enter the side A: "))
b = float(input("Enter the side B"))

c = math.sqrt(pow(a, 2) + (pow(b, 2)))

print(f"Side C: {round(c, 2)}cm")
```
