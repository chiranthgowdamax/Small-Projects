```python
num1 = float(input("Enter a number: "))
operator = input("Select an operator(+,*,/,//,%,-): ")
num2 = float(input("Enter another number: "))

if operator == "+":
  print(f"{num1} + {num2} = ", num1 + num2)
elif operator == "*":
  print(f"{num1} Ã— {num2} = ", num1 * num2)
elif operator == "/":
  print(f"{num1} / {num2} = ", num1 / num2)
elif operator == "//":
  print(f"{num1} // {num2} = ", num1 // num2)
elif operator == "-":
  print(f"{num1} - {num2} = ", num1 - num2)
elif operator == "%":
  print(f"{num1} % {num2} = ", num1 % num2)
else:
  print("{operator} is an invalid operator")
```
