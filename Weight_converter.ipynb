```python
weight = float(input("Enter your weight: "))
unit = input("Kilograms or Pounds (K or L): ")

if unit == "K":
  weight = weight * 2.205
  unit = "lbs"
  print(f"You are {round(weight,2)}{unit}")
elif unit == "L":
  weight = weight / 2.205
  unit = "kgs."
  print(f"You are {round(weight,2)}{unit}")
else:
  print(f"{unit} is invalid")
```
