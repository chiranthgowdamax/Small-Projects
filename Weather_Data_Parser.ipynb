```python
raw_string = input("Enter the City Name, Temperature and Weather Condition. Ex. City:Temperature:Weather Condition: ")

string = raw_string.strip()
parts = string.split(':')

city = parts[0]
temperature = int(parts[1])
weather = parts[2]
Fahrenheit = (temperature * 9/5)+32

city_code = city[0:3].upper()
weather = weather.lower()

print(f"Code: {city_code} | Fahrenheit: {Fahrenheit:.1f}f | Weather: {weather}")
```
